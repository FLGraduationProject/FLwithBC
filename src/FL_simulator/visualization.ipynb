```python
import matplotlib.pyplot as plt
import pickle5 as pickle
```

```python
from matplotlib import rcParams
rcParams.update({'figure.autolayout': True})
```

```python
baseline1 = 0.84
```

```python
## Load pickle
with open("../../result/data/R200_B0.3_T4_no_rep.pkl","rb") as fr:
    data = pickle.load(fr)

for i in range(50):
    plt.plot(list(data.values())[i])

plt.axhline(y=baseline1, ls='--', c='r')
plt.ylim(0,1.0)
plt.savefig("cifar10_no_rep_byz30_4T.png")
plt.show()

clientIDs = list(data.keys())

global_acc_sum = [0 for _ in range(60)]
global_count = [0 for _ in range(60)]

for clientID in clientIDs[15:]:
    for i in range(len(data[clientID])):
        global_acc_sum[i] += data[clientID][i]
        global_count[i] += 1

global_acc = [sum/count if count !=0 else 0 for (sum, count) in zip(global_acc_sum, global_count)]

global_acc_byz_no_rep = global_acc[:global_acc.index(0)]

plt.plot(global_acc_byz_no_rep)
plt.axhline(y=baseline1, ls='--', c='r')
plt.ylim(0,1.0)

plt.show()
```
